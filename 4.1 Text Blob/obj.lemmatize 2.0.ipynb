```python
from textblob import Word


word = input("Enter a word to lemmatize: ")

word_obj = Word(word)

lemmatized_noun = word_obj.lemmatize('n')  # Noun
lemmatized_verb = word_obj.lemmatize('v')  # Verb
lemmatized_adj = word_obj.lemmatize('a')   # Adjective
lemmatized_adv = word_obj.lemmatize('r')   # Adverb


print(f"Noun Lemma: {lemmatized_noun}")
print(f"Verb Lemma: {lemmatized_verb}")
print(f"Adjective Lemma: {lemmatized_adj}")
print(f"Adverb Lemma: {lemmatized_adv}")
```
