```python
from textblob import Word


word = input("Enter a word to find synonyms: ")

word_obj = Word(word)

synsets = word_obj.synsets

synonyms = set()
print('Synonyms found:')
for synset in synsets:
    for i in synset.lemma_names():
        if i != word:
            synonyms.add(i)

if synonyms:
    for synonym in synonyms:
        print(synonym)
else:
    print("No synonyms found.")
```
