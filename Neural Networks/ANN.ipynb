```python
import tensorflow as tf
import numpy as np

# Example data: [Age, BMI, Blood Pressure, Cholesterol Level]
data = np.array([
    [25, 22, 120, 180],  # Young, healthy
    [25, 28, 135, 190],  # Middle-aged, slightly overweight
    [35, 32, 140, 220],  # Obese
    [50, 30, 150, 250],  # Older, at risk
    [60, 35, 160, 260],  # High-risk
    [55, 29, 145, 240],  # Older, moderate risk
    [20, 25, 130, 200]   # Healthy middle-aged
])

labels = np.array([0, 0, 1, 1, 1, 1, 0])
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[4], activation='sigmoid')  # 4 input features and binary output
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(data, labels, epochs=40)
input_data = np.array([[55, 29, 175, 290]])
prediction = model.predict(input_data)
print("Prediction:", prediction[0][0])
if prediction[0][0] > 0.5:
    print("The model predicts that the patient has the disease.")
else:
    print("The model predicts that the patient does not have the disease.")
```
