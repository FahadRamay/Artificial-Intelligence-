```python
from collections import deque

# Graph representation as a dictionary
graph = {
    '30': ['20', '40'],
    '20': ['15', '25'],
    '40': ['35', '50'],
    '15': ['5', '18'],
    '50': ['45', '60'],
    
    '25': [],
    '35': [],
    '5': [],
    '18': [],
    '45': [],
    '60': []
}
start = '30'  # Starting node
# Initialize visited set and queue
visited = set()
queue = deque([start])

# BFS algorithm
while queue:
    node = queue.pop()  # Dequeue a node from the front of the queue
    
    if node not in visited:
        print(node)  # Process the node (in this case, we're just printing it)
        visited.add(node)  # Mark the node as visited
        
        # Enqueue all adjacent nodes that haven't been visited
        for neighbor in graph[node]:
            if neighbor not in visited:
                queue.append(neighbor)
```
