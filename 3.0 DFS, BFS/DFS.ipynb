```python
from collections import deque

graph={
    'A':['B','C'],
    'B':['D', 'E'],
'C':[],
    'D':[],
    'E':[]
}

start='A'
stack=deque([start])
visited=set()
while stack:
    node=(stack.pop())  
    if node not in visited:
        print(node)
        visited.add(node)
        for i in reversed(graph[node]):
            if i not in visited:
                stack.append(i)
    

```
