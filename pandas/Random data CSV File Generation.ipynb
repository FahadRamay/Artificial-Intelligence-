```python
import pandas as pd
import random

# Sample data
names = ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ian', 'Jack', 
         'Karen', 'Leo', 'Mia', 'Nina', 'Omar', 'Paul', 'Quinn', 'Rachel', 'Steve', 'Tom']
cities = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose']

# Generate sample data
data = {
    'ID': [i for i in range(1, 21)],
    'Name': random.sample(names, 20),  # Randomly sample from the names
    'Age': [random.randint(18, 60) for _ in range(20)],  # Random ages between 18 and 60
    'City': random.choices(cities, k=20)  # Randomly choose cities
}

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv('sample_data.csv', index=False)

# Output DataFrame for viewing
print(df)
```
