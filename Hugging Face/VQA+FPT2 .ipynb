```python
from PIL import Image
from transformers import pipeline


vqa_pipe = pipeline("visual-question-answering", model="dandelin/vilt-b32-finetuned-vqa")
image = Image.open("b.jpg")
question = "what is in this image?"
vqa_result = vqa_pipe(image, question, top_k=1)
answer = vqa_result[0]['answer']
print(answer)


text_generator = pipeline("text-generation", model="gpt2")

# Create a prompt using the answer from the VQA model
story_prompt = f"Once upon a time, there was a mysterious scene in an image. The image showed: {answer}. From this, an amazing story unfolds..."

# Generate a story based on the answer
generated_story = text_generator(story_prompt, max_length=150, num_return_sequences=1)


print(generated_story[0]['generated_text'])
```
