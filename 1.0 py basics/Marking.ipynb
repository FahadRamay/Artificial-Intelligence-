```python
from sentence_transformers import SentenceTransformer, util

# Load a pre-trained embedding model (runs locally after download)
model = SentenceTransformer('all-MiniLM-L6-v2')


qa = {
    "Define Artificial Intelligence": "Artificial Intelligence is the simulation of human intelligence processes by machines, especially computer systems.",
    "What is Machine Learning?": "Machine Learning is a subset of AI that enables systems to learn from data and improve performance without explicit programming.",
    "Explain CPU": "A CPU is the central processing unit of a computer that performs instructions and controls operations."
}

for question, correct_answer in qa.items():
    print(question)
    user_answer = input("Your Answer: ").strip()

    # Encode answers
    correct_embedding = model.encode(correct_answer, convert_to_tensor=True)
    user_embedding = model.encode(user_answer, convert_to_tensor=True)

    # Calculate cosine similarity
    similarity = util.cos_sim(user_embedding, correct_embedding).item()


    if similarity > 0.85:
        marks = 1
        result = " Correct"
    elif similarity > 0.65:
        marks = 0.5
        result = " Partially Correct"
    else:
        marks = 0
        result = " Wrong"

    print(f"Similarity: {similarity:.2f}")
    print(f"Your Answer: {user_answer}")
    print(f"Reference Answer: {correct_answer}")
    print(f"Result: {result}, Marks: {marks}\n")
```
