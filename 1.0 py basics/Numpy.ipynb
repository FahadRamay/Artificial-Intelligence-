```python
import numpy as np
ones = np.ones((3, 2)) 
zeros = np.zeros((3, 2))


result = np.dot(ones.flatten(), zeros.flatten())
print(result)
```

```python
import numpy as np
arr=np.array([2,4,5,6,10,16])
arr2=np.array([4,6,7,8,9,7])
print(np.add(arr,arr2))
print(np.mean(arr))
print("max number is :" ,np.max(arr))
print("min number is :" ,np.min(arr))
print(np.dot(arr,arr2))
```

```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6,7]
y=[100,200,400,300,500,600]
plt.title('My 1st graph')
plt.xlabel('life in years')
plt.ylabel('cost')
plt.plot(x,y, marker='o', linewidth='3', color='r',linestyle='--', label='x vs y')
plt.legend()
plt.fill_between(x,y, color='yellow')
plt.grid()
```
