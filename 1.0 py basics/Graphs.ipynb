```python
import numpy as np
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,3,4,5,6]
plt.plot(x,y, linestyle='--', marker='s',linewidth='3', color='red', label='Direction along X-axis and y-axis')
plt.title("Graph of Robot Localization")
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.legend()
plt.fill_between(x,y, color='lightgreen')

plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3, 5])
y=np.array([2,0,1,5])
plt.plot(x, y, linestyle='--',marker='o', linewidth='2',label='x vs y',color ='lightgreen')
plt.show
plt.title("X vs Y Plot")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")

plt.grid(True)
plt.fill_between(x, y)
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=[2,3,4,5]
y=[3,4,6,7]
plt.plot(x,y, linewidth='2',linestyle='--',color='r', label='x vs y', marker='s')
plt.legend()
plt.grid(True)
plt.show
plt.title('graph')
plt.xlabel("x axis")
plt.ylabel('y axis')
plt.fill_between(x,y, color='lightgreen')
```

```python
import matplotlib.pyplot as plt

# Sample data
data = [12, 15, 14, 10, 18, 20, 22, 25, 30]

plt.boxplot(data)

plt.title("Box Plot Example")
plt.ylabel("Values")

plt.show()
```

```python

import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 7]

plt.scatter(x, y)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")

plt.show()
```

```python
import matplotlib.pyplot as plt

data1 = [12, 15, 14, 10, 18]
data2 = [22, 25, 30, 28, 24]

plt.boxplot([data1, data2])

plt.title("Multiple Box Plot")
plt.xticks([1, 2], ["Group 1", "Group 2"])

plt.show()
```

```python
import matplotlib.pyplot as plt

time = [1, 2, 3, 4, 5]
values = [10, 15, 13, 17, 20]

plt.plot(time, values)

plt.xlabel("Time")
plt.ylabel("Values")
plt.title("Time Series Plot")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 15, 20, 20]

plt.step(x, y)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Step Plot")

plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]

plt.bar(categories, values)

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")

plt.show()
```

```python
plt.barh(categories, values)
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [12, 15, 13, 15, 16, 18, 20, 20, 22, 25]

plt.hist(data)

plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")

plt.show()
```
