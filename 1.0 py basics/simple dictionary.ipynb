```python

person_dict = {
    "Alice": 30,
    "Bob": 25,
    "Charlie": 35,
    "David": 28,
    "Eve": 32
}

print("Age of Alice:", person_dict["Alice"])
print("Age of Bob:", person_dict["Bob"])

person_dict["Frank"] = 40

person_dict["David"] = 29


del person_dict["Eve"]

print("Updated dictionary:")
for i, j in person_dict.items():
    print(i,j)
```

```python

countries = {
    "pakistan": "Islamabad",
    "india": "New Delhi",
    "china": "Beijing",
    "france": "Paris",
    "germany": "Berlin",
    "japan": "Tokyo",
    "canada": "Ottawa",
    "australia": "Canberra"
}


user_input = input("Enter a country name: ").strip().lower()


if user_input in countries:
    print("The capital of", user_input.title(), "is", countries[user_input])
else:
    print("Sorry, country not found in the dictionary.")
```
