```python
from langdetect import detect
from textblob import TextBlob

text = "میں کامسیٹس یونیورسٹی میں مصنوعی ذہانت کا طالب علم ہوں۔"
blob=TextBlob(text)
language=detect(blob.raw)
print(language)



```
