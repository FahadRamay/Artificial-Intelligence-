```python
import cv2
import numpy as np

# Load an image
image = cv2.imread('a.jpg',0)

# Apply a binary threshold
threshold_value = 254 # Threshold value
max_value =  150      # Maximum value to use with the THRESH_BINARY thresholding

_, binary_image = cv2.threshold(image, threshold_value, max_value, cv2.THRESH_BINARY)

# Display the binary image
cv2.imshow('Binary Image', binary_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
