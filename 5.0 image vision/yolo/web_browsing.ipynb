```python
from ultralytics import YOLO
import cv2
import webbrowser
model = YOLO("yolov8n.pt")
cap = cv2.VideoCapture(0)
phone_opened = False  
while True:
    _, frame = cap.read()
    if frame is None:
        break
        
    results = model(frame, verbose=False)
    # Get class names detected
    for result in results:
        for box in result.boxes:
            cls_id = int(box.cls[0])  # Class index box.cls is pytourch tensor of shape 1
            label = model.names[cls_id]  # Class name

            if label.lower() in ["cell phone", "mobile phone", "phone"] and not phone_opened:
                webbrowser.open("https://www.apple.com/iphone/")  
                phone_opened = True

    annotated = results[0].plot()
    cv2.imshow("YOLO Live", annotated)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
