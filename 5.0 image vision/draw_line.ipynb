```python
import cv2
import numpy as np
image = cv2.imread('a.jpg')
start_point = (200, 50)  # (x, y)
end_point = (400, 50)  # (x, y)
color = (255, 0, 0)     # BGR color for the line (here, green)
thickness = 2           # Thickness of the line (in pixels)

# Draw the line on the image
image_with_line = cv2.line(image, start_point, end_point, color, thickness)

# Display the image with the line
cv2.imshow('Image with Line', image_with_line)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
