```python
import cv2

# Load the pre-trained Haar Cascade classifier for cat face detection
cat_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalcatface.xml')

# Load an image
image_path = 'catt.jpg'
image = cv2.imread(image_path)

# Check if the image was loaded successfully
if image is None:
    print(f"Error: Unable to load image at path")
else:
    # Convert the image to grayscale (required for detection)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Perform face detection
    cat_faces = cat_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Draw rectangles around the detected faces
    for (x, y, w, h) in cat_faces:
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Display the image with the detected faces
    cv2.imshow('Detected Cat Faces', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

```
