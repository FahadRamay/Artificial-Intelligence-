```python
import cv2
import numpy as np
# Load an image
image = cv2.imread('a.jpg')
# Define the rectangle parameters (top-left corner and bottom-right corner)
start_point = (200, 50)  # (x, y)
end_point = (300, 300)  # (x, y)
color = (0, 255, 0)     # RGB color for the rectangle (here, blue)
thickness = -1           # Thickness of the rectangle border (in pixels)
# Draw the rectangle on the image
image_with_rectangle = cv2.rectangle(image, start_point, end_point, color, thickness)
# Display the image with the rectangle
cv2.imshow('Image with Rectangle', image_with_rectangle)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
