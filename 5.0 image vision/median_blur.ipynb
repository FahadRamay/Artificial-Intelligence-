```python
import cv2
import numpy as np

# Load an image
image = cv2.imread('b.jpg')

kernel_size = 9

# Apply median blur
blurred_image = cv2.medianBlur(image, kernel_size)

# Display the blurred image
cv2.imshow('Blurred Image', blurred_image)
cv2.imshow('Image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
