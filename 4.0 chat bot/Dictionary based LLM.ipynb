```python
responses = {
    "what is your name": "Fahad Amin",
    "what is your job": "Lecturer in AI",
    "where do you live": "Sahiwal"
}

while True:
    query = input("Please enter your query: ")
    query_lower = query.lower()

    if query_lower in ["exit", "quit", "bye"]:
        print("Bot: Goodbye!")
        break

    print(query_lower)

    response = responses.get(query_lower, "Sorry, I don't know the answer to that.")
    print(response)
```

```python
personal_info = {
    "name": "Fahad Amin",
    "job": "Lecturer in AI",
    "city": "Sahiwal"
}

while True:
    user_input = input("You: ").lower()

    if user_input in ["exit", "quit", "bye"]:
        print("Bot: Goodbye!")
        break

    found = False

    for key, value in personal_info.items():
        if key in user_input:
            print("Bot:", value)
            found = True
            break

    if not found:
        print("Bot: No answer found")
```
