```python
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error
import numpy as np


Experience = np.array([[1], [3], [5], [7], [9]])
Salary = np.array([70000, 75000, 80000, 85000, 90000])

#  (e.g., 'constant', 'optimal', 'invscaling', 'adaptive')
model = SGDRegressor(learning_rate='constant', eta0=0.01, max_iter=1000)
model.fit(Experience, Salary)

prediction = model.predict([[8]])
print("Prediction for 8 years of experience:", prediction[0])

predictions = model.predict(Experience)
print("Predictions for training data:", predictions)

loss = mean_squared_error(Salary, predictions)
print("Mean Squared Error (Loss):", loss)
```
