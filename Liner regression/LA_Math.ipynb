```python
import numpy as np
Experience = np.array([1, 3, 5, 7, 9]) 
Salary = np.array([70000, 75000, 80000, 85000, 90000])  
mean_X = np.mean(Experience)
mean_y = np.mean(Salary)

numerator = np.sum((Experience - mean_X) * (Salary - mean_y))
denominator = np.sum((Experience - mean_X) ** 2)

m = numerator / denominator
b = mean_y - m * mean_X

print("Intercept (bias):", b)
print("Slope (coefficient):", m)

new_experience = 8  
prediction = b + m * new_experience  

# Print the prediction
print("Prediction for 8 years of experience:", prediction)
```
