```python
from sklearn.linear_model import LinearRegression
import numpy as np

features = np.array([
    [2, 28, 16, 1, 0],  
    [3, 30, 18, 2, 1],  
    [5, 35, 20, 3, 1],  
    [7, 40, 22, 4, 2],  
    [10, 45, 24, 5, 2]  
])

salaries = np.array([60000, 70000, 80000, 90000, 110000])

model = LinearRegression().fit(features, salaries)

new_input = np.array([[4, 32, 18, 2, 1]])
predicted_salary = model.predict(new_input)

print("Predicted Salary for input",predicted_salary[0])


print("Model Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Input features: [Years of Experience, Age, Education Level (Years), Number of Certifications, Job Role]
features = np.array([
    [2, 28, 16, 1, 0],  
    [3, 30, 18, 2, 1],  
    [5, 35, 20, 3, 1],  
    [7, 40, 22, 4, 2],  
    [10, 45, 24, 5, 2]  
])

salaries = np.array([60000, 70000, 80000, 90000, 110000])

model = LinearRegression().fit(features, salaries)

new_input = np.array([[4, 32, 18, 2, 1]])
predicted_salary = model.predict(new_input)

print("Predicted Salary for input:", predicted_salary[0])
print("Model Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)

```
