```python
from sklearn.linear_model import LinearRegression
import numpy as np

exp=np.array([[1],[2],[3],[4]])
sal=np.array([60000,65000,70000,75000])
model=LinearRegression().fit(exp,sal)
print(model.predict([[5]]))
print("model has a1 as ",model.coef_)
print("model has a0 as ", model.intercept_)
```
