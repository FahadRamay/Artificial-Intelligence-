```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Data
exp = np.array([[9],[7],[5],[3],[1]])
sal = np.array([60000,65000,70000,75000,80000])

# Train model
model = LinearRegression().fit(exp, sal)

# Predictions
print(model.predict([[11]]))
print("we have value of a0 = : ", model.intercept_)
print("we have value of a1 = : ", model.coef_)

# Create smooth line for plotting
x_line = np.linspace(exp.min(), exp.max()+2, 100).reshape(-1, 1)
y_line = model.predict(x_line)

# Plot
plt.figure(figsize=(8,5))

# Scatter plot
plt.scatter(exp, sal, label="Data Points")

# Regression line
plt.plot(x_line, y_line, label="Regression Line")

# Labels
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Linear Regression Map")

# Grid
plt.grid(True)
plt.legend()

plt.show()

```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Data
exp = np.array([[1,0],[1,1],[2,1],[3,2],[4,0]])
sal = np.array([60000,80000,90000,120000,90000])

# Train model
model = LinearRegression().fit(exp, sal)

# Create a grid for both features
x1 = np.linspace(exp[:,0].min(), exp[:,0].max(), 50)
x2 = np.linspace(exp[:,1].min(), exp[:,1].max(), 50)

X1, X2 = np.meshgrid(x1, x2)

# Flatten and combine
grid_points = np.column_stack((X1.ravel(), X2.ravel()))

# Predict salaries on the grid
Y_pred = model.predict(grid_points).reshape(X1.shape)

# === Plot 3D Surface ===
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

# Scatter original data
ax.scatter(exp[:,0], exp[:,1], sal, color='red', s=60, label="Training Data")

# Regression surface
ax.plot_surface(X1, X2, Y_pred, alpha=0.5, cmap='viridis')

# Labels
ax.set_xlabel("Experience Feature 1")
ax.set_ylabel("Experience Feature 2")
ax.set_zlabel("Salary")
ax.set_title("3D Regression Surface (Two-Feature Linear Regression)")

plt.legend()
plt.show()

```
