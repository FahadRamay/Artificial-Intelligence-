```python
from sklearn.linear_model import LogisticRegression
import numpy as np
DNA_features = np.array([
    [0.1, 0.3, 0.5],   # Person 1
    [0.2, 0.1, 0.4],   # Person 2
    [0.8, 0.7, 0.9],   # Person 3
    [0.9, 0.8, 0.95],  # Person 4
    [0.4, 0.5, 0.45]   # Person 5
])
# Labels (0 = No Disease, 1 = Has Disease)
Disease_labels = np.array([0, 0, 1, 1, 0])
# Train Logistic Regression model
model = LogisticRegression().fit(DNA_features, Disease_labels)

# Predict for a new person’s DNA features
new_person_dna = np.array([[0.85, 0.9, 0.88]])  # example DNA features
prediction = model.predict(new_person_dna)
probability = model.predict_proba(new_person_dna)

print("Prediction:", "Disease" if prediction[0] == 1 else "No Disease")
print("Probability of Disease:", probability[0][1])
```
