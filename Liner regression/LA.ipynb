```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt


Experience = np.array([[1], [3], [5], [7], [9]])
Salary = np.array([70000, 75000, 80000, 85000, 90000])


model = LinearRegression().fit(Experience, Salary)


prediction = model.predict([[8]])
print("Prediction for 8 years of experience:", prediction[0])
```

```python
print("Model Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Data
Experience = np.array([[1], [3], [5], [7], [9]])
Salary = np.array([70000, 65000, 80000, 85000, 90000])

# Train model
model = LinearRegression().fit(Experience, Salary)

# Prediction
prediction = model.predict([[8]])
print("Prediction for 8 years of experience:", prediction[0])

# Plotting
plt.scatter(Experience, Salary)                   # Actual data
plt.plot(Experience, model.predict(Experience))   # Regression line

# Predicted point for 8 years
plt.scatter(8, prediction, s=100)

# Labels and title
plt.xlabel("Experience (Years)")
plt.ylabel("Salary")
plt.title("Linear Regression: Experience vs Salary")

# Show grid
plt.grid(True)

# Show plot
plt.show()
```
